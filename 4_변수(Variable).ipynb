## 2. 리스트를 복사하고자 할 때

```python
a=[1,2,3]
b=a
b
```

```python
id(b)
```

```python
id(a)
```

```python
a is b # 변수 a, 변수 b는 같은 객체를 가르킴
```

```python
# a에 있는 값 변경
a[0]=4
a
```

```python
b # a[0]을 변경했는데 b[0]도 변경됨
```

### b변수에 a를 복사할 때 값만 복사하고 다른 주소를 가리키도록 만드는 방법
- [:]이용하기

```python
a=[1,2,3]
b=a[:]
b
```

```python
id(a)
```

```python
id(b)
```

```python
a[-1]=-3
a
```

```python
b
```

- copy 모듈 이용하기

```python
from copy import copy
a=[1,2,3]
b=copy(a)
```

```python
b is a
```

```python
a[0]=5
a
```

```python
b
```

- copy() 함수 사용하기

```python
a=[1,2,3]
b=a.copy()
```

```python
id(a)
```

```python
id(b)
```

```python
a is b
```

```python
a[0]= -1
a
```

```python
b
```
