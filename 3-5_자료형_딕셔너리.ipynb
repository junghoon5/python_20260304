## 5.딕셔너리 관련 함수

### keys() : key 리스트 만들기

```python
dic={'name': '성정훈', 'phone' :'10112345678', 'birth':'1111'}
dic.keys()
```

```python
list(dic.keys())
```

```python
for key in dic.keys():
    print(key)
```

### values() : Value 리스트 만들기

```python
dic.values()
```

```python
for value in dic.values():
    print(value)
```

### items() : key, calue 쌍 얻기

```python
dic.items()
```

### clear() : key, value 쌍 모두 지우기

```python
dic
```

```python
dic.clear()
```
