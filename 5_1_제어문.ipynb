## 3. 다양한 조건을 판단하는 elif

```python
# if/else로만 표현
pocket=['paper', 'cellphone']
card= True

if'money' in pocket:
    print('택시 타고 가라')
else:
    if card:
        print('택시 타고 가라')
    else:
        ('걸어가라')
```

```python
# elif를 사용
pocket=['paper', 'cellphone']
card= True

if'money' in pocket:
    print('택시 타고 가라')
elif card:
    print('택시 타고 가라')
else:
    print('걸어 가라')
```

```python
# if문 한 줄로 작성
if 'money' in pocket: pass
else: print('카드를 꺼내라')
```

## 4.조건부 표현식

```python
# if문
score=90

if score>= 60:
    message='success'
else:
    message='failure'
print(message)
```

```python
message='success' if score>=60 else 'failure'
message
```

```python
# 조건부 표현식(conditional expression)
score=50
message='success' if score>=60 else 'failure'
message
```
